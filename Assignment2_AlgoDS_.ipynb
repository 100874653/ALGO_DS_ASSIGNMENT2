```python
#JONATHAN ANANTHAKRISHNAN
#ID: 100874653

import pygame
import time
import random

# Initializing pygame
pygame.init()

# Setting up display
WIDTH = 800
HEIGHT = 600
WIN = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Merge Sort Visualization")

# Coloring
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

# setting up sound effects
swap_sound = pygame.mixer.Sound('swap_sound.wav')  

def draw_array(array, colors):
    WIN.fill(WHITE)
    block_width = WIDTH // len(array)
    for i, val in enumerate(array):
        pygame.draw.rect(WIN, colors[i], (i * block_width, HEIGHT - val, block_width, val))
    pygame.display.update()

def merge_sort(array, left, right):
    if left < right:
        middle = (left + right) // 2
        merge_sort(array, left, middle)
        merge_sort(array, middle + 1, right)
        merge(array, left, middle, right)

def merge(array, left, middle, right):
    left_half = array[left:middle + 1]
    right_half = array[middle + 1:right + 1]

    i = j = 0
    k = left

    while i < len(left_half) and j < len(right_half):
        if left_half[i] <= right_half[j]:
            array[k] = left_half[i]
            i += 1
        else:
            array[k] = right_half[j]
            j += 1
            # Play swap sound effect
            swap_sound.play()
            time.sleep(0.2)  # Adjust delay according to your preference
            draw_array(array, [GREEN if idx == k else BLUE for idx in range(len(array))])
        k += 1

    while i < len(left_half):
        array[k] = left_half[i]
        i += 1
        k += 1

    while j < len(right_half):
        array[k] = right_half[j]
        j += 1
        k += 1
    draw_array(array, [GREEN if left <= idx <= right else WHITE for idx in range(len(array))])

def main():
    array = [random.randint(50, HEIGHT - 50) for _ in range(50)]
    draw_array(array, [WHITE] * len(array))
    merge_sort(array, 0, len(array) - 1)
    draw_array(array, [GREEN] * len(array))
    pygame.quit()

if __name__ == "__main__":
    main()
```

```python
#JONATHAN ANANTHAKRISHNAN
#ID:100874653
#QUICK SORT
import pygame
import time
import random

# Initialize pygame
pygame.init()

# Set up display
WIDTH = 800
HEIGHT = 600
WIN = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Quick Sort Visualization")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

# Set up sound effects
swap_sound = pygame.mixer.Sound('swap_sound.wav') 

def draw_array(array, colors):
    WIN.fill(WHITE)
    block_width = WIDTH // len(array)
    for i, val in enumerate(array):
        pygame.draw.rect(WIN, colors[i], (i * block_width, HEIGHT - val, block_width, val))
    pygame.display.update()

def partition(array, low, high):
    pivot = array[high]
    i = low - 1
    for j in range(low, high):
        if array[j] <= pivot:
            i += 1
            array[i], array[j] = array[j], array[i]
            # Play swap sound effect
            swap_sound.play()
            time.sleep(0.2)  # Adjust delay according to your preference
            draw_array(array, [GREEN if idx == i or idx == j else BLUE for idx in range(len(array))])
    array[i + 1], array[high] = array[high], array[i + 1]
    # Play swap sound effect
    swap_sound.play()
    time.sleep(0.2)  # Adjust delay according to your preference
    draw_array(array, [GREEN if idx == i + 1 else BLUE for idx in range(len(array))])
    return i + 1

def quick_sort(array, low, high):
    if low < high:
        pi = partition(array, low, high)
        quick_sort(array, low, pi - 1)
        quick_sort(array, pi + 1, high)

def main():
    array = [random.randint(50, HEIGHT - 50) for _ in range(50)]
    draw_array(array, [WHITE] * len(array))
    quick_sort(array, 0, len(array) - 1)
    draw_array(array, [GREEN] * len(array))
    pygame.quit()

if __name__ == "__main__":
    main()
```
